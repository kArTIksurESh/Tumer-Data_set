```python
import numpy as np
import pandas as pd
data=pd.read_csv("/content/KNNAlgorithmDataset.csv")
data
```

```python
data.isna().sum()
```

```python
data_upt=pd.get_dummies(data,{'diagnosis'})  #used to encoding
data_upt
```

```python
cnt=data.groupby('diagnosis') ['diagnosis'].count()
cnt
```

```python
del data_upt['id']
```

```python
del data_upt['diagnosis_M']
```

```python
del data_upt['Unnamed: 32']
```

```python
x=data_upt.iloc[:,:-1].values
y=data_upt.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30)
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
a=confusion_matrix(y_test,y_pred)
score=accuracy_score(y_test,y_pred)
score
```
